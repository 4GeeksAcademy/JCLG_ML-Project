# Transformación de num_dias_desde_ultima_campana

El análisis mostró que aproximadamente el 96% de los registros corresponden a clientes nunca contactados previamente, lo que evidencia que el valor 999 no representa un número real de días, sino una codificación artificial de ausencia de historial.

Mantener 999 como valor numérico habría distorsionado la interpretación del modelo, al tratar una condición categórica (“nunca contactado”) como si fuera una magnitud continua extrema.

La transformación aplicada:

Separó correctamente la dimensión conceptual en:

una variable binaria (nunca_contactado_antes),

y una variable numérica válida solo cuando existe historial real.

Preserva la semántica del negocio.

Reduce el riesgo de sesgo estructural.

Mejora la calidad de la información que recibirá el modelo.

En términos de ingeniería de datos, este paso convierte una variable mal codificada en una representación coherente, interpretable y adecuada para modelado predictivo.

```python
import pandas as pd

X_A_train_scaled = pd.DataFrame(X_A_train_scaled, columns=X_A_train.columns, index=X_A_train.index)
X_A_test_scaled = pd.DataFrame(X_A_test_scaled, columns=X_A_test.columns, index=X_A_test.index)

X_B_train_scaled = pd.DataFrame(X_B_train_scaled, columns=X_B_train.columns, index=X_B_train.index)
X_B_test_scaled = pd.DataFrame(X_B_test_scaled, columns=X_B_test.columns, index=X_B_test.index)
```

```python
# =========================
# 0) IMPORTS (clasificación)
# =========================
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# ===========================================
# 1) SPLIT TRAIN -> (TRAIN + VALIDACIÓN)
#    (Usamos SOLO train para comparar modelos)
# ===========================================

# --- Modelo A ---
X_A_tr, X_A_val, y_A_tr, y_A_val = train_test_split(
    X_A_train_scaled, y_A_train,
    test_size=0.2, random_state=42, stratify=y_A_train
)

# --- Modelo B ---
X_B_tr, X_B_val, y_B_tr, y_B_val = train_test_split(
    X_B_train_scaled, y_B_train,
    test_size=0.2, random_state=42, stratify=y_B_train
)

# =========================
# 2) CREAR "CAJAS NEGRAS"
# =========================
knn_A = KNeighborsClassifier()
dt_A  = DecisionTreeClassifier(random_state=42)
lr_A  = LogisticRegression(max_iter=2000, random_state=42)
hgb_A = HistGradientBoostingClassifier(random_state=42)

knn_B = KNeighborsClassifier()
dt_B  = DecisionTreeClassifier(random_state=42)
lr_B  = LogisticRegression(max_iter=2000, random_state=42)
hgb_B = HistGradientBoostingClassifier(random_state=42)

# =========================
# 3) ENTRENAR
# =========================
knn_A.fit(X_A_tr, y_A_tr)
dt_A.fit(X_A_tr, y_A_tr)
lr_A.fit(X_A_tr, y_A_tr)
hgb_A.fit(X_A_tr, y_A_tr)

knn_B.fit(X_B_tr, y_B_tr)
dt_B.fit(X_B_tr, y_B_tr)
lr_B.fit(X_B_tr, y_B_tr)
hgb_B.fit(X_B_tr, y_B_tr)

# ===========================================
# 4) EVALUAR EN VALIDACIÓN Y SELECCIONAR MEJOR
# ===========================================

def evaluar_modelo(nombre, modelo, X_val, y_val):
    preds = modelo.predict(X_val)
    acc = accuracy_score(y_val, preds)
    cm = confusion_matrix(y_val, preds)
    acc_por_clase = cm.diagonal() / cm.sum(axis=1)
    print(f"\n{nombre}")
    print("Accuracy total:", acc)
    print("Accuracy por clase:", acc_por_clase)
    return acc, acc_por_clase, cm

print("========== RESULTADOS VALIDACIÓN - MODELO A ==========")
acc_knn_A, accpc_knn_A, cm_knn_A = evaluar_modelo("KNN (A)", knn_A, X_A_val, y_A_val)
acc_dt_A,  accpc_dt_A,  cm_dt_A  = evaluar_modelo("Árbol (A)", dt_A, X_A_val, y_A_val)
acc_lr_A,  accpc_lr_A,  cm_lr_A  = evaluar_modelo("Regresión Logística (A)", lr_A, X_A_val, y_A_val)
acc_hgb_A, accpc_hgb_A, cm_hgb_A = evaluar_modelo("HistGradientBoosting (A)", hgb_A, X_A_val, y_A_val)

print("\n========== RESULTADOS VALIDACIÓN - MODELO B ==========")
acc_knn_B, accpc_knn_B, cm_knn_B = evaluar_modelo("KNN (B)", knn_B, X_B_val, y_B_val)
acc_dt_B,  accpc_dt_B,  cm_dt_B  = evaluar_modelo("Árbol (B)", dt_B, X_B_val, y_B_val)
acc_lr_B,  accpc_lr_B,  cm_lr_B  = evaluar_modelo("Regresión Logística (B)", lr_B, X_B_val, y_B_val)
acc_hgb_B, accpc_hgb_B, cm_hgb_B = evaluar_modelo("HistGradientBoosting (B)", hgb_B, X_B_val, y_B_val)

# (Opcional) Selección automática por accuracy total
resultados_A = {
    "KNN": acc_knn_A,
    "DecisionTree": acc_dt_A,
    "LogisticRegression": acc_lr_A,
    "HistGradientBoosting": acc_hgb_A
}
mejor_A = max(resultados_A, key=resultados_A.get)
print("\nMejor modelo A por accuracy:", mejor_A, "=>", resultados_A[mejor_A])

resultados_B = {
    "KNN": acc_knn_B,
    "DecisionTree": acc_dt_B,
    "LogisticRegression": acc_lr_B,
    "HistGradientBoosting": acc_hgb_B
}
mejor_B = max(resultados_B, key=resultados_B.get)
print("Mejor modelo B por accuracy:", mejor_B, "=>", resultados_B[mejor_B])
```
